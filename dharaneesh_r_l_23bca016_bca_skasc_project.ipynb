```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('gym.csv')
print(data.head())
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Variables', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Workout_Type', y='Resting_BPM', data=data, palette='muted')
plt.title('Resting BPM Distribution by Workout Type', fontsize=16)
plt.xlabel('Workout Type', fontsize=12)
plt.ylabel('Resting BPM', fontsize=12)
plt.xticks(rotation=45)
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Water_Intake (liters)', y='BMI', hue='Gender', data=data, palette='cool', s=100)
plt.title('Water Intake vs BMI (Colored by Gender)', fontsize=16)
plt.xlabel('Water Intake (liters)', fontsize=12)
plt.ylabel('BMI', fontsize=12)
plt.legend(title='Gender')
plt.show()
```

```python
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Creating a cluster heatmap
plt.figure(figsize=(12, 8))
sns.clustermap(
    numeric_data.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5,
    figsize=(10, 8),
    dendrogram_ratio=(0.1, 0.2),
    cbar_kws={"label": "Correlation Coefficient"}
)
plt.title('Cluster Heatmap of Numeric Variables', pad=20)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data, x='Calories_Burned', shade=True, label='Calories Burned', color='blue')
sns.kdeplot(data=data, x='BMI', shade=True, label='BMI', color='green')
sns.kdeplot(data=data, x='Fat_Percentage', shade=True, label='Fat Percentage', color='red')
plt.title('KDE Plot for Calories Burned, BMI, and Fat Percentage', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(title='Variable', fontsize=10)
plt.show()
```

```python
from statsmodels.distributions.empirical_distribution import ECDF


variables = ['Calories_Burned', 'BMI', 'Fat_Percentage']

plt.figure(figsize=(10, 6))
for var in variables:
    ecdf = ECDF(data[var])
    plt.plot(ecdf.x, ecdf.y, label=var)

plt.title('ECDF Plot for Selected Variables', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('ECDF', fontsize=12)
plt.legend(title='Variable', fontsize=10)
plt.grid(True)
plt.show()
```

```python
numeric_columns = data_cleaned.select_dtypes(include=['float64', 'int64']).columns
outlier_info = {}
for column in numeric_columns:
    Q1 = data_cleaned[column].quantile(0.25)
    Q3 = data_cleaned[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = ((data_cleaned[column] < lower_bound) | (data_cleaned[column] > upper_bound)).sum()
    data_cleaned = data_cleaned[~((data_cleaned[column] < lower_bound) | (data_cleaned[column] > upper_bound))]
    outlier_info[column] = outliers
```

```python
# Reloading the new uploaded dataset
data = pd.read_csv('/mnt/data/gym.csv')

# Step 1: Check and handle missing values
data_cleaned = data.dropna()

# Step 2: Remove duplicate rows
data_cleaned = data_cleaned.drop_duplicates()

# Step 3: Detect and remove outliers using the IQR method for numeric columns
numeric_columns = data_cleaned.select_dtypes(include=['float64', 'int64']).columns
for column in numeric_columns:
    Q1 = data_cleaned[column].quantile(0.25)
    Q3 = data_cleaned[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    data_cleaned = data_cleaned[~((data_cleaned[column] < lower_bound) | (data_cleaned[column] > upper_bound))]

cleaned_file_path = '/mnt/data/cleaned_gym_dataset.csv'
data_cleaned.to_csv(cleaned_file_path, index=False)

cleaned_file_path
```
